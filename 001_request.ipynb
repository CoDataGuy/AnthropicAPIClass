```python
# Create API Client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-5"
```

```python
# Make a request
message = client.messages.create (
    model=model,
    max_tokens = 1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)

print("\n--- Full response object ---")
print(message)
print("\n--- Text ---")
print(message.content[0].text)
```
